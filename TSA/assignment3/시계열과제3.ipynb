```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ar_2(n,p):
    L=[]
    x1,x2=0,0
    wt=np.random.normal(0,1,n)
    for t in range(n):
        xt=p*x2+wt[t]
        L.append(xt)
        x2=x1
        x1=xt
    return L
```

```python
ar2=ar_2(100,-0.9)
```

```python
def sacf(x):
    L=[]
    n=len(x)
    mu=np.mean(x)
    for h in range(0,20):
        Lh=[]
        for t in range(n-h):
            ac=(x[t+h]-mu)*(x[t]-mu)
            Lh.append(ac)
        autocov_h=sum(Lh)/n
        L.append(autocov_h)
    atcv=np.array(L)
    sac=atcv/atcv[0]
    return sac
```

```python
plt.stem(sacf(ar2))
plt.show()
```

```python
def ma_2(n,t1,t2):
    L=[]
    wt=np.random.normal(0,1,n+2)
    for t in range(2,n+2):
        xt=wt[t]+t1*wt[t-1]+t2*wt[t-2]
        L.append(xt)
    return L
```

```python
ma2=ma_2(100,0.5,-0.2)
plt.stem(sacf(ma2))
plt.show()
```

```python
def arma_21(n,p1,p2,t1):
    L=[]
    x1,x2=0,0
    wt=np.random.normal(0,1,n+1)
    for t in range(1,n+1):
        xt=p1*x1+p2*x2+wt[t]-t1*wt[t-1]
        L.append(xt)
        x2=x1
        x1=xt
    return L
```

```python
arma21=arma_21(100,0.8,-0.15,-0.3)
```

```python
arma21=arma_21(100,0.8,-0.15,-0.3)
plt.stem(sacf(arma21))
plt.show()
```
